```python
import numpy
import matplotlib.pyplot as plt
from numpy.random import normal
```

```python
### LINEAR REGRESSION ###

# make fake data
x = numpy.arange(0.0,100.0,0.1)
y = numpy.zeros(len(x))
for i in range(len(x)):
    y[i] = 1.4*x[i]**2 + 1.1*x[i] + 3000.0
#    y[i] += (normal()/6.*x[i])**2 + (normal()*10*x[i]) + 10.0*normal()
    y[i] += 700.0*normal()
    
# do linear regression
from sklearn.linear_model import LinearRegression
linreg = LinearRegression(fit_intercept=True)

linreg.fit([[t, t**2] for t in x], y)

intercept = linreg.intercept_
coef = linreg.coef_

print "--------------"
print "fit coefficients:"
print intercept, coef
print "--------------"

prediction = linreg.predict([[t, t**2] for t in x])

# plot prediction and original data
plt.figure(figsize=(16,6))
plt.subplot(1,2,1)
plot(x, y, 'or', label='data')
plot(x, prediction, '--k', lw='5', label='prediction')
plt.legend(loc='upper left')

plt.subplot(1,2,2)
#plot(x, (prediction-y)**2)
catch = plt.hist((prediction-y), bins=40, label='residuals')
plt.legend(loc='upper left')
```

```python
alpha = numpy.arange(0.0,100.0,0.01)
N = len(alpha)
x1, x2 = numpy.zeros(N), numpy.zeros(N)

y = numpy.zeros(N)

# create features x1, x2 and create y
features = []
for i in range(N):
    x1[i] = 1.2*alpha[i] + 3.0*normal() + 20
    x2[i] = 0.2*alpha[i]**2 + 6.0*normal() + 20
    y[i] = 3.0*x1[i] + 0.1*x2[i] + 50.0*normal() + 1000
    features.append([x1[i],x2[i]])
    
trainFeatures = features[::2]
trainY = y[::2]
testFeatures = features[1::2]
testY = y[1::2]

x1Train = [x[0] for x in trainFeatures]
x2Train = [x[1] for x in trainFeatures]
x1Test  = [x[0] for x in testFeatures]
x2Test  = [x[1] for x in testFeatures]
    
plt.figure(figsize=(16,5))
plt.subplot(1,3,1)
plot(x1Train, x2Train, 'or', label='training')
plot(x1Test, x2Test, 'oy', label='test', alpha=0.5)
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend(loc='upper left')

plt.subplot(1,3,2)
plot(x1Train, trainY, 'or', label='training')
plot(x1Test, testY, 'oy', label='test', alpha=0.5)
plt.xlabel('x1')
plt.ylabel('y')
plt.legend(loc='upper left')

plt.subplot(1,3,3)
plot(x2Train, trainY, 'or', label='training')
plot(x2Test, testY, 'oy', label='test', alpha=0.5)
plt.xlabel('x2')
plt.ylabel('y')
plt.legend(loc='upper left')

#########################################################
# do linear regression
from sklearn.linear_model import LinearRegression
linreg = LinearRegression(fit_intercept=True)

linreg.fit(trainFeatures, trainY)

intercept = linreg.intercept_
coef = linreg.coef_

score = linreg.score(testFeatures, testY)

print "--------------"
print "fit coefficients:"
print intercept, coef
print "--------------"
print "R^2 score:"
print score
print "--------------"

prediction = linreg.predict(testFeatures)

# plot prediction and original data
plt.figure(figsize=(16,5))
plt.subplot(1,3,1)
plot(x1Test, testY, 'oy', label='test data')
plot(x1Test, prediction, '--k', lw='5', label='prediction')
plt.legend(loc='upper left')

plt.subplot(1,3,2)
plot(x2Test, testY, 'oy', label='test data')
plot(x2Test, prediction, '--k', lw='5', label='prediction')
plt.legend(loc='upper left')

plt.subplot(1,3,3)
#plot(x, (prediction-y)**2)
catch = plt.hist((prediction-testY), bins=40, label='residuals', alpha=0.8)
plt.legend(loc='upper left')
```
